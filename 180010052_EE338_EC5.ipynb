```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
from scipy import fft, ifft
import math
```

```python
t = np.linspace(-5, 5, 100)
x = np.cos(t)
def rect(t, a2):
    result = []
    a = a2/2
    for i in t:
        if(np.abs(i)<a):
            result.append(1)
        else:
            result.append(0)
    return result
plt.plot(t, x*rect(t, 15))
plt.show()

# The following figures illustrate the Heisenberg principle as discussed in the attached paper
```

## Analysis of ITC Stock Price

```python
stock = np.genfromtxt('ITC_2yr_closing.csv', delimiter=',')
stock[0] = 273.05 #error in importing
stock
```

```python
N = len(stock)
t_itc = np.linspace(0, N, N)
plt.figure(figsize=(16,8))
plt.grid(True)
plt.plot(t_itc, stock)
plt.show()
```

```python
y = fft(stock)
xf = np.linspace(0.0, 0.5, N//2)
plt.figure(figsize=(16,5))
plt.plot(xf[1::], 2.0/N * np.abs(y[1:N//2])) # 1:: index is present since we cannot plot a reliable graph without
#neglecting the dc component
plt.show()

# Thus, the stock prices are dominated by low frequencies
# Note: The frequency unit is (day)^-1
```

```python
# Proceeding to calculate the CWT
omega0 = 0.01
def psift(s):
    return np.pi**(-.25)*np.exp(-0.5*(s*xf-omega0)**2)
def ycwt(s):
    return y[:N//2]*psift(s)
plt.figure(figsize=(16,5))
plt.xlim(0, 0.1)
plt.plot(xf[1::], 2.0/N * np.abs(ycwt(1)[1::]))
plt.show()
```

```python
def stockcwt(s):
    return ifft(ycwt(s))
plt.figure(figsize=(16,8))
plt.grid(True)
plt.plot(t_itc[::2], np.abs(stockcwt(1)))
plt.plot(t_itc[::2], np.abs(stockcwt(2)))
plt.plot(t_itc[::2], np.abs(stockcwt(20)))

plt.plot(t_itc[::2], stock[::2], 'r')
plt.show()
```
